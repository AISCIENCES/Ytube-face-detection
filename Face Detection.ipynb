```python
import cv2
import face_recognition
from google.colab.patches import cv2_imshow
```

```python
elonImage = face_recognition.load_image_file('sample_data/elon-1.jpg')
elonImage = cv2.cvtColor(elonImage, cv2.COLOR_BGR2RGB)
cv2_imshow(elonImage)
```

```python
facelocation = face_recognition.face_locations(elonImage)[0]
cv2.rectangle(elonImage, (facelocation[3], facelocation[0]), (facelocation[1], facelocation[2]), (30,255,0),5)
cv2_imshow(elonImage)
```

```python
elonEncoding = face_recognition.face_encodings(elonImage)[0]
elonEncoding
```

```python
testImage = face_recognition.load_image_file('sample_data/donald-1.jpg')
testImage = cv2.cvtColor(testImage, cv2.COLOR_BGR2RGB)
facelocation = face_recognition.face_locations(testImage)[0]
cv2.rectangle(testImage, (facelocation[3], facelocation[0]), (facelocation[1], facelocation[2]), (30,255,0),5)
testEncoding = face_recognition.face_encodings(testImage)[0]
matched = face_recognition.compare_faces([elonEncoding], testEncoding)[0]
cv2_imshow(testImage)
matched
```
